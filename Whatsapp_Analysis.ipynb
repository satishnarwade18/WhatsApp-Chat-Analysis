```python
pattern = pattern = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages = re.split(pattern, data)[1:]
```

```python
df = pd.DataFrame({"user_message":messages,"message_date":dates})
```

```python
users = []
messages = []

for message in df['user_message']:
    entry = re.split(r'^(.+?):\s', message)
    #entry = re.split('([\w\W]+?):\s',message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(" ".join(entry[2:]))
    else:
        users.append('group notification')
        messages.append(entry[0])
        
df['user'] = users
df['message'] =messages

df.drop(columns=['user_message'],inplace=True)
```

```python
period = []
for hour in df[['day_name','hour']]['hour']:
    if hour == 23:
        period.append(str(hour)+"-"+str("00"))
    elif hour == 0:
        period.append(str("00")+"-"+str(hour+1))
    else:
        period.append(str(hour)+"-"+str(hour+1))
    
```

```python
df['period'] = period
```

```python
df['period']
```

```python
import seaborn as sns

# Create a pivot table that counts the messages for each combination of 'day_name' and 'period'
pivot_table = df.groupby(['day_name', 'period']).size().unstack(fill_value=0)

# Plot the heatmap
plt.figure(figsize=(20, 6))
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", cbar=True)
plt.yticks(rotation='horizontal')
plt.show()
```
